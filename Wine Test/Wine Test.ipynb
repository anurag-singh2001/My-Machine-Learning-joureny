```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('winetest.csv')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
missing_val_count_by_column = (df.isnull().sum())
print(missing_val_count_by_column[missing_val_count_by_column > 0])
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x='type',data=df)
plt.show()
```

```python
#converting categorical to numerical
df['type']=pd.get_dummies(df['type'],drop_first=True)
df
```

```python
# correlation
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),cmap='coolwarm')
plt.show()
```

```python
#ploting countplot quality 
plt.figure(figsize=(5,5))
sns.countplot(x='quality',data=df)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df.drop('quality',axis=1))
```

```python
from sklearn.model_selection import train_test_split
X=df.drop('quality',axis=1)
y=df['quality']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
#svm
from sklearn.svm import SVC 
svm = SVC(kernel='linear', random_state=0)  
svm=svm.fit(X_train, y_train)  
svm=svm.score(X_test,y_test)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
lr=lr.score(X_test,y_test)
```

```python
# random forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100)
rf.fit(X_train,y_train)
rf=rf.score(X_test,y_test)
```

```python
# knn
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
knn=knn.score(X_test,y_test)
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
dt=dt.score(X_test,y_test)
```

```python
# naive bayes
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train,y_train)
nb=nb.score(X_test,y_test)
```

```python
models = pd.DataFrame({
    'Model': ['Linear Regression', 'KNN', 'SVM', 
              'Random Forest', 'Naive Bayes', 'Decision Tree'],
    'Score': [lr,knn,svm,rf,nb,dt]})
```

```python
models.sort_values(by='Score', ascending=False)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Model',y='Score',data=models)
plt.show()
```
