# Removing Old Teams

```python
# Removing old teams
all_teams = np.unique(matches['team1'])
old_teams = ['Deccan Chargers', 'Kochi Tuskers Kerala', 'Gujarat Lions', 'Kochi Tuskers Kerala', 'Rising Pune Supergiant', 'Pune Warriors', 'Rising Pune Supergiants']
current_teams = [teams for teams in all_teams if teams not in old_teams] 
matches= matches[matches['team1'].isin(current_teams) & matches['team2'].isin(current_teams)]
matches
```

# Most Wins in IPL

```python
temp = pd.DataFrame({"Winner": matches['winner']})
count_wins = temp.value_counts()
#print(count_wins)
labels = [X[0] for X in count_wins.keys()]

bar , ax = plt.subplots(figsize=(20,12))
ax=plt.pie(x = count_wins, autopct = "%.1f%%" ,labels= labels)
plt.title("Most Wins in IPL", fontsize = 17)
plt.show()
```

```python
sns.countplot(matches['winner'][matches['eliminator']=='Y'],data=matches)
plt.title("Most wins in Eliminator",fontsize=17)
plt.xticks(rotation=90)
plt.show()
```

```python
teams = matches['toss_winner'].unique()

decision_making = pd.DataFrame([],columns=['Toss Winner','Decision','Times'])

for id, element in enumerate(teams):
    
    temp_bat = matches[(matches['toss_winner']==element) & (matches['toss_decision']=='bat')]
    temp_field = matches[(matches['toss_winner']==element) & (matches['toss_decision']=='field')]
    
    decision_making = decision_making.append({'Toss Winner':element, 'Decision':'bat',
                                              'Times':temp_bat['toss_winner'].count()}, ignore_index=True)
    
    decision_making = decision_making.append({'Toss Winner':element, 'Decision':'field', 
                                              'Times':temp_field['toss_winner'].count()}, ignore_index=True)
```

```python
decision_making
```

# Decision of Bat or Bowl after winning toss

```python
sns.catplot(x= "Toss Winner", y = 'Times', hue = "Decision" , data = decision_making, kind = 'bar', height= 5, aspect=2 )

plt.xticks(rotation = 90)
plt.title("Toss Decision of Teams")
plt.xlabel("IPL Teams")
plt.ylabel(" Toss Decision ")
plt.show()
```

```python
teams = matches['toss_winner'].unique()

decision_making = pd.DataFrame([],columns=['Toss Winner','Decision','Match wins after winning Toss'])

for id, element in enumerate(teams):
    
    temp_bat = matches[(matches['toss_winner']==element) & (matches['toss_decision']=='bat') & (matches['toss_winner']==matches['winner'])]
    temp_field = matches[(matches['toss_winner']==element) & (matches['toss_decision']=='field') & (matches['toss_winner']==matches['winner'])]
    
    decision_making = decision_making.append({'Toss Winner':element, 'Decision':'bat',
                                              'Match wins after winning Toss':temp_bat['toss_winner'].count()}, ignore_index=True)
    
    decision_making = decision_making.append({'Toss Winner':element, 'Decision':'field', 
                                              'Match wins after winning Toss':temp_field['toss_winner'].count()}, ignore_index=True)
```

```python
decision_making
```

# Match wins after winning Toss

```python
sns.catplot(x= "Toss Winner", y = 'Match wins after winning Toss', hue = "Decision" , data = decision_making, kind = 'bar', height= 5, aspect=2 )

plt.xticks(rotation = 90)
plt.title("Toss Decision of Teams")
plt.xlabel("IPL Teams")
plt.ylabel(" Toss Decision ")
plt.show()
```

```python
# matches['method']!='NaN'
teams = matches['winner'].unique()
DL = pd.DataFrame([],columns=['Winner','Win with the help of D/L Method'])
for id, element in enumerate(teams):
    dl=matches[(matches['method']!='NaN')&(matches['winner']==element)]
    DL=DL.append({'Winner':element, 'Win with the help of D/L Method':dl['winner'].count()}, ignore_index=True)
    
DL    
```

```python
sns.catplot(x= "Winner", y = 'Win with the help of D/L Method' , data = DL, kind = 'bar', height= 5, aspect=2 )

plt.xticks(rotation = 90)
plt.title("Team Win using D/L method")
plt.xlabel("IPL Teams")
plt.ylabel("Wins")
plt.show()
```

```python
teams = matches['venue'].unique()
DL = pd.DataFrame([],columns=['Venue','Number of D/L happens at this venue'])
for id, element in enumerate(teams):
    dl=matches[(matches['method']!='NaN')&(matches['venue']==element)]
    DL=DL.append({'Venue':element, 'Number of D/L happens at this venue':dl['venue'].count()}, ignore_index=True)
    
DL    
```

```python
sns.catplot(x= "Venue", y = 'Number of D/L happens at this venue' , data = DL, kind = 'bar', height= 5, aspect=2 )

plt.xticks(rotation = 90)
plt.title("Number Of D/L happened at particular Venue")
plt.xlabel("Venue")
plt.ylabel("Matches")
plt.show()
```

```python
batsmanruns=deliveries.groupby(['batsman'])['total_runs'].sum().reset_index()
batsmanruns = batsmanruns.sort_values(["total_runs"], ascending=False)
batsmanruns
```

```python
topbat=batsmanruns[:10]
sns.catplot(x= "batsman", y = 'total_runs' , data = topbat, kind = 'bar', height= 5, aspect=2 )
plt.xticks(rotation = 90)
plt.title("Most Runs scored by Batsman")
plt.xlabel("Batsman")
plt.ylabel("Total Run")
plt.show()
```

```python
wickets=deliveries.groupby(['bowler'])['is_wicket'].sum().reset_index()
wickets = wickets.sort_values(["is_wicket"], ascending=False)
wickets
```

```python
topbolwer=wickets[:10]
sns.catplot(x= "bowler", y = 'is_wicket' , data = topbolwer, kind = 'bar', height= 5, aspect=2 )
plt.xticks(rotation = 90)
plt.title("Most Wicket by Bolwer")
plt.xlabel("Bolwer")
plt.ylabel("Total Wicket")
plt.show()
```
